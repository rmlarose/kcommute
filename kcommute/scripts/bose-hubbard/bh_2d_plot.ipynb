```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("bh_graph-2D-grid-nonpbc-qubitnodes_Lx-5_Ly-5_U-100_enc-unary_d-4.csv")
```

```python
k = df["i"].to_numpy() + 1 # k values in original program are off by one
rhats = df["rhat"].to_numpy()
i_max = np.argmax(rhats)
kstar = k[i_max]
print(kstar)
print(np.max(rhats) / np.min(rhats))
```

```python
kvals = df["i"].to_numpy() + 1
rhats = df["rhat"].to_numpy()
groups = df["ngroups"].to_numpy()

fig, ax = plt.subplots(2, 1, sharex=True)

color = 'tab:orange'
ax[0].set_ylabel(r'$\hat{R}$', color=color)
ax[0].plot(kvals, rhats, color=color)
ax[0].tick_params(axis='y', labelcolor=color)

color = 'tab:blue'
ax[1].set_ylabel('Groups', color=color)  # we already handled the x-label with ax1
ax[1].plot(kvals, groups, color=color)
ax[1].tick_params(axis='y', labelcolor=color)
ax[1].set_xlabel('k')

fig.tight_layout()  # otherwise the right y-label is slightly clipped

plt.savefig("bose_hubbard_2d.pdf")
```
