# Plot results

```python
import json

import matplotlib.pyplot as plt
```

```python
fnames = ["16.out", "20.out", "24.out", "28.out", "32.out"]
```

```python
plt.rcParams.update({"font.family": "serif", "font.size": 12})

for fname in fnames:
    data = json.load(open(fname, "rb"))
    plt.plot(
        [int(n) for n in data["rhat_vs_k"].keys()],
        data["rhat_vs_k"].values(),
        "--s",
        ms=7,
        mec="black",
        alpha=0.85,
        label=f"$n = {data['nqubits']}$",
    )
plt.xlabel("$k$")
plt.ylabel(r"$\hat{R}$")
plt.legend();
```
